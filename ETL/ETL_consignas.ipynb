2. Los valores nulos de los campos revenue, budget deben ser reemplazados por 0.

4. De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
#Paso de formato str a formato date time la solumna released date
dfmovies['release_date'] = pd.to_datetime(dfmovies['release_date'], errors='coerce')
```
